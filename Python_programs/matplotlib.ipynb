```python
import matplotlib.pyplot as plt
import numpy as np
```

# line plot:

```python
x=[6,2,8,4,7]
y=[2,5,7,8,3]
#line plot
plt.plot(x,y)
```

```python
x1=[1,2,3,4,5]
y1=[4,5,6,7,8]
plt.xlabel('x values')
plt.ylabel('y values')
plt.title('Line plot')
plt.plot(x1,y1)
```

# Bar plot:

```python
x=[6,2,8,4,7]
y=[2,5,7,8,3]
#bar plot
plt.bar(x,y)
```
